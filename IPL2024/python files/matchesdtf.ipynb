```python
import pandas as pd
import json
```

**MATCH SUMMARY**

```python
with open('matches.json') as f:
    data = json.load(f)
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
df_match.rename({'Scorecard': 'match_id'}, axis =1, inplace = True)
df_match.head()
```

**BATTING SUMMARY**

```python
with open('batting.json') as b:
    data = json.load(b)

    all_records=[]
    for record in data:
        all_records.extend(record['battingSummary'])

df_batting = pd.DataFrame(all_records)

df_batting.drop('strikerate',axis=1,inplace = True)
df_batting.rename({'strikrate': 'strikerate'}, axis =1, inplace = True)
df_batting.tail()
```

```python
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x: "not_out" if x=='not out' else "out")
df_batting.head(50)
```

```python

runs= df_batting['runs'].astype(int)
balls = df_batting['balls'].astype(int)
calcstrikerate = runs/balls*100

df_batting['calcstrikerate'] = round(calcstrikerate,2).astype(str)

df_batting.tail()
```

```python
df_batting.drop('dismissal',axis=1,inplace = True)
df_batting.head(50)
```

```python
df_batting.drop('strikerate',axis=1,inplace = True)
df_batting.head(50)
```

```python
df_batting.rename({'calcstrikerate': 'sr'}, axis =1, inplace = True)
```

```python
df_batting.head(500)
```

```python
df_match.head()
```

```python
df_batting.to_csv('batting.csv',index=False)
```

```python
df_match.to_csv('match.csv', index = False)
```

**PLAYER INFO**

```python
with open('players.json') as f:
    data = json.load(f)
```

```python
df_players = pd.DataFrame(data)

print(df_players.shape)
df_players.head(10)
```

```python
df_players['name'] = df_players['name'].apply(lambda x: x.replace('â€', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('†', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('\xa0', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace("(c)", ''))
df_players.head(10)
```

```python
df_players.to_csv('playerswithoutimage.csv', index = False)
```
